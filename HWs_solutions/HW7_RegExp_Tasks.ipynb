#### **Задание 14**

Получите как можно больше чисел из этого набора.

Ответ: `['4', '8.0', '+16', '-16', '-23.42', '3.14e15', '23e-42', '100.000.000', '-3.099e-734.149']`


```python
number_pattern = r'[+-]?(?:\d+)' + r'(?:\.\d+)*'
float_ending = r'(?:\.\d+)?'
exponential_part = fr'(?:[eE]{number_pattern})?'
number_pattern = number_pattern + float_ending + exponential_part

pattern = number_pattern 
```

```python
text = '4 8.0 +15 -16 -23.42 3.14e15 23e-42 100.000.000 -3.099e-734.149'
result = re.findall(pattern = pattern, string = text)
print(result)
print(f'Получилось {len(result)} чисел, ожидается 9')
```

#### **Задание 15**

Получите все валидные номера телефона из 'data/phones.txt'.

Ответ: `['"7(911) 345-34-56"', '"7(923)355-56-53"', '"+7(923)355-56-53"', '"8(988)245 45 32"', '"88005553535"', '"+7 921 445 43 22"']`


```python
sep = r'[\s-]?'
phone_code =  r'(?:\+7|[78])' + r'[\s\(]?' + r'\d{3}' + '[\s\)]?'
pattern = phone_code + r'\s?' +  r'\d{3}' + sep + r'\d{2}' + sep + r'\d{2}'
pattern = fr'\"{pattern}\"'
```

```python
with open('data/phones.txt', mode='r') as f:
    text = f.read()
result = re.findall(pattern = pattern, string = text)
print(result)
print(f'Получилось {len(result)} номеров, ожидается 6')
```

```python
# Я бы лучше сделал repl отдельной функцией,
# чтобы строка не была миллион символов в длину.
# Но пишу тут с lambda просто чтобы показать как еще можно
def transsalt(text: str) -> str:
    pattern = r'([уеыаоэяиюёУЕЫАОЭЯИЮЁ])'
    salty_text = re.sub(pattern = pattern, 
                        repl = lambda vowel: vowel.group(0) + 'с' + vowel.group(0).lower(), 
                        string = text)
    return salty_text
```

```python
# Поскольку тексты песен обычно не по 200 Гб,
# то обрабатывать не лету в целом не обязательно
def extract_chords(path: str) -> set:
    pattern =r"\b[A-G][m]?\b"
    with open(path, mode='r', encoding='utf-8') as file:
        text = file.read()
    chords = re.findall(pattern=pattern, string=text)
    return set(chords)
```
